### Task 7 - Mini problem

```python
def add_price(price_list, price):
    if price.isdigit():
        price_list.append(int(price))
        print("Price added successfully.")
    else:
        print("Please enter a valid number.")


def get_average_price(price_list):
    if len(price_list) > 0:
        return sum(price_list) / len(price_list)
    else:
        return None


def get_max_price(price_list):
    if len(price_list) > 0:
        return max(price_list)
    else:
        return None


price_list = []

while True:
    menu_items = input(
        "\n1 - Add price\n"
        "2 - Show average price\n"
        "3 - Show max price\n"
        "q - Quit\n"
        "Enter your choice: "
    )

    if menu_items == '1':
        price = input("Please enter the price: ")
        add_price(price_list, price)

    elif menu_items == '2':
        avg = get_average_price(price_list)
        if avg is not None:
            print("Average price:", avg)
        else:
            print("Price list is empty")

    elif menu_items == '3':
        maximum = get_max_price(price_list)
        if maximum is not None:
            print("Max price:", maximum)
        else:
            print("Price list is empty")

    elif menu_items == 'q':
        print("Exiting...")
        break

    else:
        print("Invalid choice")
```
