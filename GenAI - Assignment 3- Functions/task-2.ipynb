#### Task 2 Recursive Function : Factorial Utility

```python
def factorial(n):
    if n < 0 :
        return 'The given number is negative'
    else:
        if n == 1 or n == 0:
            return 1
        else:
            return n * factorial(n-1)

print(factorial(5))
print(factorial(0))
print(factorial(-3))
```
