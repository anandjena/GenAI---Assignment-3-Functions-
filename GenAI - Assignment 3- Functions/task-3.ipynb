### Task 3 : Lambda Gst Calculator

```python
gst = lambda price : price + (price * 18/100)
print(gst(100))
```

```python
## Calculating final price after discount and adding gst

final_price = lambda price,discount,gst : (price - price * discount / 100) + ((price - price * discount / 100) * gst / 100)
print(final_price(100,10,18))
```
